```python
df = pd.read_csv(r"/content/pop_2023 datasheet.csv", encoding='Windows-1252')
df.head()
```

```python
df.drop(['Population per km2  of Arable Land'],axis=1,inplace=True)
df.drop(['Projected Deaths per 100,000\nDue to Temperature Change,\n2040-2059\nAnnual Avg.d'],axis=1,inplace=True)
df.drop(['Internally Displaced Persons Due to Disasters'],axis=1,inplace=True)
df.drop(['Urban Population Living\nin Slum Households\n(%)'],axis=1,inplace=True)
df.drop(['Demand for Family Planning Satisfied by Modern Methods (%)f'],axis=1,inplace=True)
df.drop(['Family Plann Married(Using All Methods (%))'],axis=1,inplace=True)
df.drop(['Family Plann Married(Using Modern Methods (%))'],axis=1,inplace=True)
df.drop(['Ratio of Female to Male Un- employment Rates'],axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
df['GNI per Capita, PPPc']=df['GNI per Capita, PPPc'].fillna(df['GNI per Capita, PPPc'].mean())
df['Urban Population\n(%)']=df['Urban Population\n(%)'].fillna(df['Urban Population\n(%)'].mean())
df['Moderate/ Severe Food Insecurity (%)']=df['Moderate/ Severe Food Insecurity (%)'].fillna(df['Moderate/ Severe Food Insecurity (%)'].mean())

df.info()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Visualize the distribution of numerical features
plt.figure(figsize=(12, 8))
sns.histplot(data=df, x='Population (millions)\nmid-2023', bins=20, kde=True)
plt.title('Distribution of Population mid-2023')
plt.show()

```

```python

# Explore relationships between variables
plt.figure(figsize=(12, 8))
sns.pairplot(df[['Population (millions)\nmid-2023', 'Total Fertility Ratea', 'Life Expectancy at Birth (Years)Total']])
plt.title('Pairplot of Selected Variables')
plt.show()

```

```python
# Correlation heatmap to understand feature relationships
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Explore categorical variables
plt.figure(figsize=(30, 14))
sns.countplot(data=df, x='Country')
plt.xticks(rotation=90)
plt.title('Count of Countries')
plt.show()
```

```python

# Boxplot for various features
plt.figure(figsize=(20, 12))
sns.boxplot(data=df.drop(columns=['Country']))
plt.title('Boxplot of Various Features')
plt.xticks(rotation=45)
plt.show()

```

```python
# Violin plot for selected features
plt.figure(figsize=(14, 30))
sns.violinplot(x='Life Expectancy at Birth (Years)Total', y='Country', data=df)
plt.title('Violin Plot of Life Expectancy by Country')
plt.show()

```

```python
# Scatter plot matrix for numerical features
plt.figure(figsize=(20, 16))
sns.pairplot(df.drop(columns=['Country']), diag_kind='kde')
plt.suptitle('Scatter Plot Matrix', y=1.02)
plt.show()

```

```python
# Bar plot for selected features
plt.figure(figsize=(50, 16))
sns.barplot(x='Country', y='Urban Population\n(%)', data=df)
plt.title('Bar Plot of Urban Population by Country')
plt.xticks(rotation=45)
plt.show()
```

```python

# Line plot for time-series data (if applicable)
# Example: Population growth over the years
time_series_columns = ['Population (millions)\nmid-2023', 'Population (millions)mid-2035', 'Population (millions)mid-2050']
df_time_series = df[['Country'] + time_series_columns]
df_time_series.set_index('Country').T.plot(figsize=(25, 10), marker='o')
plt.title('Population Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Population (millions)')
plt.show()
```
